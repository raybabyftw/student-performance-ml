```python
# 05-analysis.ipynb

# 1. Imports
import pandas as pd
import numpy as np
import pickle
import tensorflow as tf
from tensorflow import keras
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import os
from IPython.display import display

# Ensure results dir exists
os.makedirs("results", exist_ok=True)

# 2. Load test data
test_df = pd.read_csv("processed_data/test.csv")
X_test = test_df.drop(columns=["G3_binary"])
y_test = test_df["G3_binary"]

# Wide & Deep splits
cat_cols = [col for col in X_test.columns if any(prefix in col for prefix in 
    ['school_', 'sex_', 'address_', 'famsize_', 'Pstatus_', 'Mjob_', 'Fjob_', 
     'reason_', 'guardian_', 'schoolsup_', 'famsup_', 'paid_', 'activities_', 
     'nursery_', 'higher_', 'internet_', 'romantic_'])]
num_cols = [col for col in X_test.columns if col not in cat_cols]
X_test_wide = X_test[cat_cols]
X_test_deep = X_test[num_cols]

# 3. Load models
with open("models/baseline_model.pkl", "rb") as f:
    baseline_model = pickle.load(f)
seq_model = keras.models.load_model("models/sequential_model.h5")
wide_deep_model = keras.models.load_model("models/wide_deep_model.h5")

# 4. Load tuning results
seq_results_df = pd.read_csv("results/sequential_hyperparameter_results.csv")
wide_deep_results_df = pd.read_csv("results/widendeep_hyperparameter_results.csv")

print("Sequential Model Hyperparameter Tuning Results:")
display(seq_results_df)

print("\nWide & Deep Model Hyperparameter Tuning Results:")
display(wide_deep_results_df)

# 5. Evaluate models
def evaluate_model(name, model, probs, preds):
    acc = accuracy_score(y_test, preds)
    prec = precision_score(y_test, preds, zero_division=0)
    rec = recall_score(y_test, preds, zero_division=0)
    f1 = f1_score(y_test, preds, zero_division=0)
    auc = roc_auc_score(y_test, probs)
    
    print(f"\n{name} Performance:")
    print(classification_report(y_test, preds))
    print(f"ROC-AUC Score: {auc:.4f}")
    
    cm = confusion_matrix(y_test, preds)
    plt.figure(figsize=(4,3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=[0,1], yticklabels=[0,1])
    plt.title(f"{name} Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()
    
    return {
        "Model": name,
        "Accuracy": round(acc, 4),
        "Precision": round(prec, 4),
        "Recall": round(rec, 4),
        "F1-Score": round(f1, 4),
        "ROC-AUC": round(auc, 4)
    }

# Baseline
baseline_probs = baseline_model.predict_proba(X_test)[:, 1]
baseline_preds = (baseline_probs > 0.5).astype(int)
baseline_results = evaluate_model("Baseline (LogReg)", baseline_model, baseline_probs, baseline_preds)

# Sequential
seq_probs = seq_model.predict(X_test).ravel()
seq_preds = (seq_probs > 0.5).astype(int)
seq_results = evaluate_model("Sequential Model", seq_model, seq_probs, seq_preds)

# Wide & Deep
wd_probs = wide_deep_model.predict([X_test_wide, X_test_deep]).ravel()
wd_preds = (wd_probs > 0.5).astype(int)
wd_results = evaluate_model("Wide & Deep Model", wide_deep_model, wd_probs, wd_preds)

# 6. Comparison table
final_results = pd.DataFrame([baseline_results, seq_results, wd_results])
print("\nFinal Model Comparison Table:")
display(final_results)

# Save final results
final_results.to_csv("results/final_model_comparison.csv", index=False)
print("✅ Final results saved to results/final_model_comparison.csv")

# 7. Plot F1-score comparison
plt.figure(figsize=(6,4))
sns.barplot(x="Model", y="F1-Score", data=final_results)
plt.title("Model Comparison by F1-Score")
plt.ylabel("F1-Score")
plt.show()
```
