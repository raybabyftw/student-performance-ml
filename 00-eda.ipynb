```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("processed_data/train.csv")
df.head()
```

```python
print("Shape:", df.shape)
print("\nData types:\n", df.dtypes)
print("\nMissing values:\n", df.isnull().sum())
```

```python
df.describe().T
```

```python
sns.countplot(x='G3_binary', data=df)
plt.title('Distribution of Target: High Performers (G3  15)')
plt.xlabel('G3_binary (0 = Not High Performer, 1 = High Performer)')
plt.ylabel('Count')
plt.show()
```

```python
# Compute correlation matrix and select top 15 features most correlated with target
target_col = 'G3_binary'
cor_matrix = df.corr()
top_corr_features = cor_matrix[target_col].abs().sort_values(ascending=False).head(15).index

# Filter to only the top correlated features
filtered_corr = df[top_corr_features].corr()

# Plot cleaner, focused correlation matrix
plt.figure(figsize=(10, 8))
sns.set(style="white")

ax = sns.heatmap(
    filtered_corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    center=0,
    linewidths=0.5,
    cbar_kws={"shrink": 0.75}
)

plt.title("Correlation Matrix (Top 15 Features)", fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
selected_features = ['studytime', 'failures', 'absences', 'goout', 'health']
for feature in selected_features:
    plt.figure()
    sns.boxplot(x='G3_binary', y=feature, data=df)
    plt.title(f'{feature} vs. G3_binary')
    plt.show()
```

```python
sample_cols = ['studytime', 'failures', 'absences', 'G3_binary']
sns.pairplot(df[sample_cols], hue='G3_binary')
plt.show()
```
